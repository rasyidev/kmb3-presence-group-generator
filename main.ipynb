```python
import os
from zoom_presence import ZoomPresence
import pandas as pd
```

```python
folder_name = "20220913"
files = os.listdir(f"{folder_name}/")
times = [
'08:00 AM',
'08:00 AM',
'08:00 AM',
'08:00 AM',
'11:00 AM',
'08:00 AM',
'08:00 AM',
'08:00 AM'
]
ft = list(zip(files, times))

f = open(f"filename-and-timestamp/{folder_name}.txt", "a")
for i in ft:
    print(i)
    f.write(str(i))
    f.write("\n")
f.close()
```

```python
df = None
for i, (f, t) in enumerate(ft):
    zp = ZoomPresence(f"{folder_name}/{f}", t, )
    zp.gen_good_presence()
    if i == 0:
        df = zp.good_presence
    else:
        df = pd.concat([df, zp.good_presence], ignore_index=True)
df
```

```python
zp.good_presence = df
zp.save_as_excel(folder_name, zp.output_name)
```
